```python
import pandas as pd
import os
```

```python
dataset_url = 'https://archivo.datos.cdmx.gob.mx/incidentes-viales-c5.csv'
```

```python
data = pd.read_csv(dataset_url)
```

```python
data['fecha_creacion'] = pd.to_datetime(data['fecha_creacion'])
data['fecha_cierre'] = pd.to_datetime(data['fecha_cierre'])
data2 = data.loc[(data['fecha_creacion'].dt.year==2020)]
data2 = data2.reset_index(drop=True)
```

```python
def save2File(data, path, name):
    with pd.ExcelWriter(os.path.join(path,name)) as writer: # Open and create the file
        data.to_excel(writer, sheet_name='DB',index=False) # Saving the information of database
```

```python
save2File(data2,'.','MX_Road_Accidents_2020.xlsx')
```
